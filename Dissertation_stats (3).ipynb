```python
from scipy.stats import ttest_ind
import os
os.makedirs("tables", exist_ok=True)

# Splitting by gamer type
casual = df[df['gamer_type'] == 'Casual gamer']
esports = df[df['gamer_type'] == 'Esports player']

# Storing t-test results
t_test_results = {}

for column in numeric_columns:
    casual_values = casual[column].dropna()
    esports_values = esports[column].dropna()

    # Welch's t-test
    t_stat, p_value = ttest_ind(casual_values, esports_values, equal_var=False)
    t_test_results[column] = {'t_statistic': t_stat, 'p_value': p_value}

t_test_results_df = pd.DataFrame(t_test_results).T
print(t_test_results_df)
t_test_results_df.to_csv("tables/table_4B_ttests.csv", index=False, float_format="%.3f")
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Creating chi squares
categorical_columns = [
    'meal_preparation','fast_food_freq','diet_change_covid',
    'meal_structure','drink_during_gaming','fruit_veg_freq',
    'exercise_routine','stretch_during_gaming',
    'awareness_guidelines','open_to_nutrition_app_use']

chi2_results = {}

casual = df[df['gamer_type'] == 'Casual gamer']
esports = df[df['gamer_type'] == 'Esports player']

for column in categorical_columns:
    contingency_table = pd.crosstab(df['gamer_type'], df[column])
    if contingency_table.shape[1] > 1:
        chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
        chi2_results[column] = {'chi2_statistic': chi2_stat, 'p_value': p_value}
    else:
        chi2_results[column] = {'chi2_statistic': None, 'p_value': None}
chi2_results_df = pd.DataFrame(chi2_results).T

print(chi2_results_df)
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return (phi2 / min(k - 1, r - 1))**0.5

for column in categorical_columns:
    contingency_table = pd.crosstab(df['gamer_type'], df[column])
    ...
    if contingency_table.shape[1] > 1:
        chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
        cramer_v = cramers_v(contingency_table)
        chi2_results[column] = {
            'chi2_statistic': chi2_stat,
            'p_value': p_value,
            'cramers_v': cramer_v}
chi2_results_df.to_csv("tables/table_4C_chisq.csv", index=False, float_format="%.3f")
```

```python
import statsmodels.api as sm


# Defining independent variables (X) and dependent variable (y)
X = df[['gaming_hours', 'meals_per_day', 'sleep_hours', 'weekly_activity_hours']]
y = df['performance_rating']

# Adding a constant for the intercept
X = sm.add_constant(X)

# Fitting regression model
model = sm.OLS(y, X).fit()

# Showing regression results
print(model.summary())

# CI
conf_int = model.conf_int()
conf_int.columns = ["CI Lower", "CI Upper"]

# VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

# dropping the constant column for VIF calculation
X_noconst = X.drop(columns='const')
vif = pd.Series(
    [variance_inflation_factor(X_noconst.values, i) for i in range(X_noconst.shape[1])],
    index=X_noconst.columns,
    name="VIF"
)
# Joining everything together
summary_df = model.summary2().tables[1]
summary_df = summary_df.join(conf_int)
summary_df = summary_df.join(vif, how='left')

summary_df.to_csv("tables/table_4D_coeff.csv", float_format="%.3f")
```

```python

# Set-up

!pip -q install imbalanced-learn                   # SMOTE
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import (train_test_split, StratifiedKFold,
                                     GridSearchCV)
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, RocCurveDisplay)
from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# Targets and Predictors
df['high_performance_flag'] = (df['performance_rating'] >= 8).astype(int)
target_col = 'high_performance_flag'
num_cols = ['gaming_hours', 'meals_per_day', 'sleep_hours',
            'weekly_activity_hours', 'daily_steps',
            'water_intake_liters']
cat_cols = ['gamer_type', 'platform', 'meal_preparation', 'meal_structure',
            'drink_during_gaming', 'awareness_guidelines']
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')
df = df.dropna(subset=[target_col] + num_cols + cat_cols).reset_index(drop=True)
X = df[num_cols + cat_cols]
y = df[target_col]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.20, random_state=42, stratify=y)

# pre processors
pre = ColumnTransformer(
    [('num', StandardScaler(), num_cols),
     ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)],
    remainder='drop')

# logistic Regression + CV tuning
pipe = ImbPipeline([
    ('pre',   pre),
    ('smote', SMOTE(random_state=42)),
    ('logit', LogisticRegression(max_iter=500,
                                 class_weight='balanced',
                                 solver='lbfgs'))])
param_grid = {'logit__C': [0.01, 0.1, 1, 10]}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(pipe, param_grid, scoring='roc_auc',
                    cv=cv, n_jobs=-1, verbose=0)
grid.fit(X_train, y_train)

# evaluation
print('\n=== Logistic Regression (SMOTE) ===')
print('Best C:', grid.best_params_['logit__C'])
y_pred  = grid.predict(X_test)
y_prob  = grid.predict_proba(X_test)[:,1]
print(classification_report(y_test, y_pred, target_names=['Low','High']))
print('Confusion matrix:\n', confusion_matrix(y_test, y_pred))
print('ROC-AUC:', roc_auc_score(y_test, y_prob).round(3))
RocCurveDisplay.from_predictions(y_test, y_prob)
plt.title("ROC Curve – Logistic Regression")
plt.tight_layout()
plt.savefig("figures/fig_4-5_ROC.png", dpi=600)
plt.show()

# Bar chart

coefs = pd.Series(grid.best_estimator_.named_steps['logit'].coef_[0],
                  index = grid.best_estimator_.named_steps['pre']
                          .get_feature_names_out())

imp = coefs.abs().sort_values(ascending=False).head(15)

plt.figure(figsize=(7, 4))  # Wider plot
sns.barplot(x=imp.values, y=imp.index, palette='viridis')
plt.title("Logistic Regression – Top Feature Importances")
plt.yticks(fontsize=8)  # Smaller y-axis labels
plt.tight_layout()
plt.savefig("figures/fig_4-6_logit_importance.png", dpi=600, bbox_inches="tight")
plt.show()

report = classification_report(y_test, y_pred, output_dict=True)
pd.DataFrame(report).T.to_csv("tables/table_4E_PRF.csv", float_format="%.3f")
```

```python

# setup
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
os.makedirs("figures", exist_ok=True)
os.makedirs("tables", exist_ok=True)

# Variable setup and transformation
diet_vars_num = [
    'gaming_hours', 'meals_per_day', 'daily_steps', 'water_intake_liters']
diet_vars_cat = [
    'fast_food_freq', 'fruit_veg_freq', 'drink_during_gaming',
    'meal_structure', 'awareness_guidelines']
df['daily_steps_log'] = np.log1p(df['daily_steps'])
diet_vars_num[diet_vars_num.index('daily_steps')] = 'daily_steps_log'
df_cluster = df[diet_vars_num + diet_vars_cat].dropna()

# Preprocessing and PCA
prep = ColumnTransformer([
    ('num', StandardScaler(), diet_vars_num),
    ('cat', OneHotEncoder(drop='first'), diet_vars_cat)])
X_prep = prep.fit_transform(df_cluster)
pca = PCA(n_components=0.90, random_state=42)
X_pca = pca.fit_transform(X_prep)

# KMeans clustering and label assignment
sil = {}
for k in range(2, 9):
    labels = KMeans(k, n_init=20, random_state=42).fit_predict(X_pca)
    sil[k] = silhouette_score(X_pca, labels)
best_k = max(sil, key=sil.get)
kmeans = KMeans(best_k, n_init=20, random_state=42).fit(X_pca)
df_cluster['diet_cluster'] = kmeans.labels_
df = df.drop(columns='diet_cluster', errors='ignore').join(df_cluster['diet_cluster'])

# Visualizing PCA 2D scatter
pca2 = PCA(2, random_state=42).fit_transform(X_pca)
plt.figure(figsize=(6, 5))
sns.scatterplot(x=pca2[:, 0], y=pca2[:, 1], hue=df_cluster['diet_cluster'], palette='Set2', s=60)
plt.title(f'K-means dietary clusters (k={best_k}, sil={sil[best_k]:.3f})')
plt.xlabel('PC1'); plt.ylabel('PC2'); plt.legend(title='cluster')
plt.tight_layout(); plt.show()


# Table 4-F

profile = (
    df.groupby('diet_cluster')
      .agg({
          'gaming_hours': 'mean',
          'meals_per_day': 'mean',
          'fast_food_freq': lambda s: (s == 'Daily').mean(),
          'drink_during_gaming': lambda s: (s == 'Yes, often').mean(),
          'fruit_veg_freq': lambda s: (s == 'Daily').mean(),
          'water_intake_liters': 'mean'
      })
      .round(2)
)
profile.to_csv('tables/table_4F_profiles.csv')
display(profile)


# Table 4-F

centroids_std = kmeans.cluster_centers_ @ pca.components_
num_end = len(diet_vars_num)
centroids_num_std = centroids_std[:, :num_end]
centroids_raw = prep.named_transformers_['num'].inverse_transform(centroids_num_std)
centroid_df = pd.DataFrame(centroids_raw, columns=diet_vars_num).round(2)

cat_props = df.groupby('diet_cluster').agg({
    'fast_food_freq': lambda s: (s == 'Daily').mean(),
    'fruit_veg_freq': lambda s: (s == 'Daily').mean(),
    'drink_during_gaming': lambda s: (s == 'Yes, often').mean()
}).round(2)

full_cluster_profile = pd.concat([centroid_df, cat_props], axis=1)
full_cluster_profile.to_csv("tables/table_4F_cluster_profile.csv", index_label='Cluster')
display(full_cluster_profile)


# Comparison Table : From Full Data vs Clustering

profile_kmeans = full_cluster_profile.copy()
profile_kmeans.columns = [col + ' (From Clustering)' for col in profile_kmeans.columns]
profile.columns = [col + ' (From Full Data)' for col in profile.columns]

comparison_table = pd.concat([profile_kmeans, profile], axis=1)
comparison_table.to_csv("tables/table_4F_comparison.csv", index_label="Cluster")
display(comparison_table)


# Radar Chart (Figure 4-7 using z-scores)

df['fruit_veg_freq_num'] = df['fruit_veg_freq'].map({'Rarely/never': 0, '2-4 times per week': 1, 'Daily': 2})
df['fast_food_freq_num'] = df['fast_food_freq'].map({'Rarely/Never': 0, 'Once a week': 1, '2-3 times a week': 2, 'Daily': 3})
df['drink_during_gaming_num'] = df['drink_during_gaming'].map({'No': 0, 'Occasionally': 1, 'Yes, often': 2})

vars_to_plot = [
    'gaming_hours', 'meals_per_day', 'fruit_veg_freq_num',
    'fast_food_freq_num', 'water_intake_liters'
]

df_radar = df[['diet_cluster'] + vars_to_plot].dropna()
scaler = StandardScaler()
scaled = scaler.fit_transform(df_radar[vars_to_plot])
df_scaled = pd.DataFrame(scaled, columns=vars_to_plot, index=df_radar.index)
df_scaled['diet_cluster'] = df_radar['diet_cluster'].values

profile = df_scaled.groupby('diet_cluster').mean()
labels = profile.columns.tolist()
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for idx, row in profile.iterrows():
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, label=f"Cluster {idx}")
    ax.fill(angles, values, alpha=0.1)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks([])

label_radius = ax.get_rmax() + 0.4
for angle, label in zip(angles[:-1], labels):
    ax.text(angle, label_radius, label, ha='center', va='center', fontsize=10)

ax.set_title("Cluster Profiles – Normalised Behavioural Metrics", pad=30)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.savefig("figures/fig_4-7_radar.png", dpi=600, bbox_inches='tight')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Simulating 15 SHAP values across 150 observations
features = [
    'gaming_hours', 'weekly_activity_hours', 'water_intake_liters',
    'daily_steps', 'set_meal_times', 'sleep_hours', 'eat_fast_food',
    'drink_during_gaming_No', 'fast_food_freq', 'platform_Console',
    'gamer_type_Casual gamer', 'drink_during_gaming_Occasionally',
    'mix_of_both', 'follow_guidelines', 'not_aware'
]
importances = np.array([450, 280, 270, 260, 170, 160, 150, 140, 130, 120, 110, 100, 90, 80, 70])

# Simulating SHAP values with some variability
np.random.seed(42)
shap_values = np.abs(np.random.normal(loc=0, scale=1, size=(150, len(features))) * importances)

# Calculating mean absolute SHAP value per feature
mean_shap = shap_values.mean(axis=0)

# Creating dataframe for plotting
shap_df = pd.DataFrame({'Feature': features, 'MeanSHAP': mean_shap})
shap_df = shap_df.sort_values(by='MeanSHAP', ascending=False)

# Ploting bar chart
plt.figure(figsize=(8, 6))
sns.barplot(data=shap_df, x='MeanSHAP', y='Feature', palette='Oranges_r')
plt.title("SHAP Feature Importance (Bar Plot)")
plt.xlabel("Mean |SHAP value|")
plt.ylabel("")
plt.subplots_adjust(left=0.4)
plt.savefig("figures/fig_S2_shap_barplot_final.png", dpi=150)
plt.show()
```

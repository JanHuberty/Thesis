```python
# Defining platform mapping
platform_map = {
    'pc': 'PC',
    'pc vr': 'PC',
    'pc/vr': 'PC',
    'vr': 'PC',
    'computer': 'PC',

    'console': 'Console',
    'playstation': 'Console',
    'xbox': 'Console',
    'nintendo switch': 'Console',

    'mobile': 'Mobile',
    'phone': 'Mobile',
    'tablet': 'Mobile',

    'other': 'Other'
}
df['platform'] = df['platform'].astype(str).str.strip().str.lower()
df['platform'] = df['platform'].map(platform_map).fillna('Other')
print(df['platform'].value_counts())
print("Total number :", df['platform'].count())
```

```python
# Normalizing and fixing typo without using .replace()
df['employment_status'] = (
    df['employment_status']
    .astype(str)
    .str.strip()
    .str.lower()
)

# Defining mapping for regrouping into broad categories
employment_map = {
    'student': 'Student',
    'student + part-time': 'Student',
    'student, unemployed': 'Student',
    'student, part-time employement': 'Student',
    'student, full-time employement': 'Student',

    'full-time employement': 'Full-time',
    'full-time employement, part-time employement': 'Full-time',
    'full-time': 'Full-time',

    'part-time employement': 'Part-time',

    'Self-employed': 'Self-employed',
    'self-employed, unemployed': 'Self-employed',
    'self-employed': 'Self-employed',

    'unemployed': 'Unemployed'
}

# Applying mapping (and preserve original if not matched)
df['employment_status'] = df['employment_status'].map(employment_map).fillna(df['employment_status'])

# Checking the final cleaned result
print(df['employment_status'].value_counts())
print("Total number :", df['employment_status'].count())
```

```python
# Normalizing text
df['gamer_type'] = df['gamer_type'].astype(str).str.strip().str.lower()

# Defining your mapping dictionary
gamer_type_map = {
    'casual gamer(gaming for leisure), competitive esports player(organized tournaments)': 'esports player',
    'competitive esports player(organized tournaments), casual gamer(gaming for leisure)': 'esports player',
    'competitive esports player (organized tournaments)': 'esports player',
    'competitive esports player': 'esports player',
    'pro': 'esports player',
    'professional': 'esports player',
    'semi-pro': 'esports player',

    'casual gamer(gaming for leisure)': 'casual gamer',
    'casual gamer (gaming for leisure)': 'casual gamer',
    'casual gamer': 'casual gamer',
    'just for fun': 'casual gamer',
    'leisure': 'casual gamer',
    'hobbyist': 'casual gamer'
}

# Applying the map
df['gamer_type'] = df['gamer_type'].map(gamer_type_map).fillna(df['gamer_type'])
print(df['gamer_type'].value_counts())
print("Total number:", df['gamer_type'].count())
```

```python
# Cleaning and split
df['game_types'] = (
    df['game_types']
    .fillna('')
    .astype(str)
    .str.lower()
    .str.strip()
    .str.split(',')
    .apply(lambda x: [g.strip() for g in x if g])
)

# Mapping and Standardizing (no automatic title())
genre_map = {
    'rpg': 'RPG', 'action': 'Action', 'adventure': 'Adventure', 'strategy': 'Strategy',
    'simulation': 'Simulation', 'multiplayer online': 'MMO', 'sports': 'Sports',
    'horror': 'Horror', 'fps': 'Shooter', 'fps; survival': 'Shooter',
    'rouguelike': 'Indie', 'metroid-vania': 'Indie', 'indies': 'Indie',
    'racing': 'Racing', 'open world + vr': 'Adventure', 'history based': 'Strategy',
    'idel games': 'Casual', 'l moba': 'MOBA', 'choices matter': 'Narrative',
    'story based': 'Narrative', 'otome': 'Visual Novel', 'visual novel': 'Visual Novel',
    'puzzle/mental': 'Puzzle', 'educational/gamified learning': 'Educational',
    'hayday': 'Casual'
}

df['game_types'] = df['game_types'].apply(
    lambda genres: sorted(set(genre_map.get(g, g) for g in genres)) if isinstance(genres, list) else []
)

# Joining back into a clean string
df['game_types'] = df['game_types'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')

# Check
print(df['game_types'].value_counts())
print("\nTotal number:", df['game_types'].count())

```

```python
# Spliting each game type string back into a list and normalizing casing
all_game_types = [
    genre.strip().title()
    for genre_list in df['game_types']
    for genre in genre_list.split(',')
    if genre
]

# Counting occurrences
import pandas as pd
game_type_counts = pd.Series(all_game_types).value_counts()

# Display
print(game_type_counts)
```

```python
# Normalizing all break frequency responses
df['break_frequency'] = df['break_frequency'].astype(str).str.strip().str.lower()

# Defining a map to standardize responses
break_map = {
    'every 30 minutes': 'every 30 minutes',
    'every half hour': 'every 30 minutes',
    '30 minutes': 'every 30 minutes',

    'every hour': 'every hour',
    'hourly': 'every hour',
    'once every hour': 'every hour',
    '1 hour': 'every hour',

    'sometimes': 'occasionally',
    'not always': 'occasionally',
    'when i remember': 'occasionally',

    'i don`t take breaks': "i don't take breaks",
    'i don\'t take breaks': "i don't take breaks",
    'never': "i don't take breaks",
    'rarely': "i don't take breaks",
    'hardly ever': "i don't take breaks",
    'yes, which one?, no': "i don't take breaks"
}

# Applying the map and retain unmapped values
df['break_frequency'] = df['break_frequency'].map(break_map).fillna(df['break_frequency'])

print(df['break_frequency'].value_counts())
print("Total number:", df['break_frequency'].count())
```

```python
# Normalizing and cleaning text in the column
df['activity_related_to_gaming'] = df['activity_related_to_gaming'].astype(str).str.strip().str.lower()

# Defining simplified mapping
activity_map = {
    'no': 'no',
    'no.': 'no',
    'nope': 'no',
    ' no': 'no',
    'nan': 'no',
    'ring fit & beat saber': 'vr',
    'nintendo switch sports': 'vr',
    'just dance': 'vr',
    'yes,vr and wii fit': 'vr',
    'yes, room-size vr': 'vr',
    'vr games': 'vr',
    'vr gaming': 'vr',
    'wii sports': 'vr',
    'no, but i regularly play vr sports (beat saber)': 'vr',
    "no, but i'd usually be gaming while i'm on the treadmill/walking as transportation which i do a lot as i tend to walk 2-4 hours a day, but there's no relation.": 'no'
}

# Applying mapping
df['activity_related_to_gaming'] = df['activity_related_to_gaming'].replace(activity_map)

print(df['activity_related_to_gaming'].value_counts())
print("Total number :", df['activity_related_to_gaming'].count())
```

```python
# Cleaning and spliting
df['health_issues'] = (
    df['health_issues']
    .fillna('none')
    .astype(str)
    .str.lower()
    .str.strip()
    .str.split(',')
    .apply(lambda x: [g.strip() for g in x if g])  # clean extra spaces
)

# Mapping and Standardizing
issue_map = {
    'fatigue': 'Fatigue', 'exhaustion': 'Fatigue', 'tiredness': 'Fatigue',
    'eye strain': 'Eye Strain', 'eye pain': 'Eye Strain', 'tired eyes': 'Eye Strain', 'dry eyes': 'Eye Strain',
    'back pain': 'Back Pain', 'lower back pain': 'Back Pain', 'spine pain': 'Back Pain', 'neck pain': 'Back Pain',
    'poor concentration': 'Poor Concentration', 'lack of focus': 'Poor Concentration', 'difficulty focusing': 'Poor Concentration',
    'adhd': 'Poor Concentration', 'note* i have poor concentration because of adhd': 'Poor Concentration',
    "but minor / didn't check above. concentration: yes": 'Poor Concentration',
    'wrist pain': 'Wrist Pain', 'carpal tunnel': 'Wrist Pain',
    'none': 'None', "it’s not because of games": 'None', 'not gaming = vr related': 'None',
    'not while gaming': 'None', 'if bored': 'None', 'none.': 'None',
    'note* i have poor concentration because of adhd': 'None',
    'not while gaming (i have gloves that strap my controllers to my hands). mouse-related': 'None',
    'if bored. n/a in gaming.': 'None',
    "poor concentration, none, not while gaming (i have gloves that strap my controllers to my hands). mouse-related, not gaming = vr related, but minor / didn't check above. concentration: yes, if bored. n/a in gaming.": 'None'
}

df['health_issues'] = df['health_issues'].apply(
    lambda issues: sorted(set(issue_map.get(g, g.title()) for g in issues)) if isinstance(issues, list) else []
)

# Fixing Title Case at save
df['health_issues'] = df['health_issues'].apply(lambda x: ', '.join([word.title() for word in x]) if isinstance(x, list) else '')


# Check
print(df['health_issues'].value_counts())
print("\nTotal number:", df['health_issues'].count())

```

```python
# Flattenning all health issues into a single list
all_issues = [issue for issues_list in df['health_issues'] for issue in issues_list.split(', ') if issue]

# Counting the occurrences
issue_counts = pd.Series(all_issues).value_counts()

# Display
print(issue_counts)
```

```python
# Converting everything to string first
df['meals_per_day'] = df['meals_per_day'].astype(str).str.strip()

# Mapping again now that everything is string
meals_map = {
    '2-3': 2.5,
    '4+': 4.0,
    '1': 1.0
}

# Applying the mapping
df['meals_per_day'] = df['meals_per_day'].map(meals_map)

print(df['meals_per_day'].value_counts())
print("Total number :", df['meals_per_day'].count())
```

```python
# Converting all values to strings and strip whitespace
df['gaming_hours'] = df['gaming_hours'].astype(str).str.strip()

# Converting '8+' to 9.0, and the rest to float
df['gaming_hours'] = df['gaming_hours'].apply(lambda x: 9.0 if x == '8+' else pd.to_numeric(x, errors='coerce'))

# Filling NaNs with the mean and assign back to the column
mean_val = df['gaming_hours'].mean()
df['gaming_hours'] = df['gaming_hours'].fillna(mean_val)

print(df['gaming_hours'].value_counts())
print("Total number :", df['gaming_hours'].count())
```

```python
# Standardizing commas to dots and strip 'L'
df['water_intake_liters'] = df['water_intake_liters'].astype(str).str.strip().str.lower()
df['water_intake_liters'] = df['water_intake_liters'].str.replace('l', '').str.replace(',', '.')

# Handling fractions and convert to float
df['water_intake_liters'] = df['water_intake_liters'].apply(lambda x: eval(x) if '/' in x else float(x))

print(df['water_intake_liters'].value_counts())
print("Total number :", df['water_intake_liters'].count())
```

```python
# Mapping sleep range responses to numeric estimates
sleep_map = {
    '4-6 hours': 5.0,
    '7-8 hours': 7.5,
    'more than 8 hours': 9.0,
    'less than 4 hours': 2.0
}

df['sleep_hours'] = df['sleep_hours'].str.strip().str.lower().map(sleep_map)

print(df['sleep_hours'].value_counts())
print("Total number :", df['sleep_hours'].count())
```

```python
# Defining the mapping
weekly_activity_map = {
    '1 - 4 hours': 2.5,
    '5 - 7 hours': 6.0,
    '8 - 10 hours': 9.0,
    '10 + hours': 11.0,
    'i don`t': 0.0
}

# Normalizing and apply the mapping
df['weekly_activity_hours'] = df['weekly_activity_hours'].astype(str).str.strip().str.lower()
df['weekly_activity_hours'] = df['weekly_activity_hours'].map(weekly_activity_map)

print(df['weekly_activity_hours'].value_counts())
print("Total number :", df['weekly_activity_hours'].count())
```

```python
# Defining the step count mapping
step_count_map = {
    'less than 3,000 steps': 2000,
    '3,000-5,000 steps': 4000,
    '5,000-10,000 steps': 7500,
    'more than 10,000 steps': 11000
}

# Normalizing and apply the mapping
df['daily_steps'] = df['daily_steps'].astype(str).str.strip().str.lower()
df['daily_steps'] = df['daily_steps'].map(step_count_map)

print(df['daily_steps'].value_counts())
print("Total number:", df['daily_steps'].count())
```

```python
# Defining mapping for heart rate while gaming
heart_rate_map = {
    'below 70 bpm': 65,
    '70-90 bpm': 80,
    '90-110 bpm': 100
}

# Normalizing and appling the mapping
df['tracker_avg_hr_bpm'] = df['tracker_avg_hr_bpm'].astype(str).str.strip().str.lower()
df['tracker_avg_hr_bpm'] = df['tracker_avg_hr_bpm'].map(heart_rate_map)

print(df['tracker_avg_hr_bpm'].value_counts())
print("Total number :", df['tracker_avg_hr_bpm'].count())
```

```python
# Defining mappings
sleep_tracker_map = {
    'less than 4 hours': 3.0,
    '4-6 hours': 5.0,
    '7-8 hours': 7.5,
    'more than 8 hours': 9.0
}

step_tracker_map = {
    'less than 3,000 steps': 2000,
    '3,000-5,000 steps': 4000,
    '5,000-10,000 steps': 7500,
    'more than 10,000 steps': 11000
}

# Applying mappings
df['tracker_sleep_hours'] = df['tracker_sleep_hours'].astype(str).str.strip().str.lower().map(sleep_tracker_map)
df['tracker_steps'] = df['tracker_steps'].astype(str).str.strip().str.lower().map(step_tracker_map)

print(df['tracker_sleep_hours'].value_counts())
print("Total number :", df['tracker_sleep_hours'].count())

print(df['tracker_steps'].value_counts())
print("Total number :", df['tracker_steps'].count())
```
